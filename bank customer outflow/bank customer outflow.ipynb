Из «Бета-Банка» стали уходить клиенты. Каждый месяц. Немного, но заметно. Банковские маркетологи посчитали: сохранять текущих клиентов дешевле, чем привлекать новых.

Нужно спрогнозировать, уйдёт клиент из банка в ближайшее время или нет. Вам предоставлены исторические данные о поведении клиентов и расторжении договоров с банком. 

Постройте модель с предельно большим значением *F1*-меры. Чтобы сдать проект успешно, нужно довести метрику до 0.59. Проверьте *F1*-меру на тестовой выборке самостоятельно.

Дополнительно измеряйте *AUC-ROC*, сравнивайте её значение с *F1*-мерой.

### Разделение выборки

## Исследование задачи

### Логистическая регрессия

```python
# Создадим функцию вычисления метрик
def metrics (model, features_valid, target_valid):

# Вычислим метрику F1
    predicted_valid = model.predict(features_valid)
    print('F1-мера:', f1_score(target_valid, predicted_valid))

# Посчитаем площадь под ROC-кривой — AUC-ROC
    probabilities_valid = model.predict_proba(features_valid)
    auc_roc = roc_auc_score(target_valid, probabilities_valid[:, 1])
    print('AUC-ROC:', auc_roc)
```

```python
model_log = LogisticRegression(random_state=12345) 

# Обучаем модель на тренировочной выборке
model_log.fit(features_train, target_train) 

# Расчитываем качество модели на валидационной выборке
result = model_log.score(features_valid, target_valid) 

metrics(model_log, features_valid, target_valid)

# Вычислим метрику F1
#predicted_valid = model_log.predict(features_valid)
#print('F1-мера:', f1_score(target_valid, predicted_valid))

# Посчитаем площадь под ROC-кривой — AUC-ROC
#probabilities_valid = model_log.predict_proba(features_valid)
#auc_roc = roc_auc_score(target_valid, probabilities_valid[:, 1])
#print('AUC-ROC:', auc_roc)
```

### Модель - Деревья решений

```python
best_model_tree = None
best_result_tree = 0

# Создаем цикл для выбора глубины дерева решений
for depth in range(1, 15):
    model_tree = DecisionTreeClassifier(random_state=12345, max_depth=depth) 
    
# Обучаем модель на тренировочной выборке
    model_tree.fit(features_train, target_train)

# Получаем предсказание модели
    predictions = model_tree.predict(features_train) 

# Расчитываем качество модели на валидационной выборке
    result = model_tree.score(features_valid, target_valid) 
    if result > best_result_tree:

# Сохраняем наилучшую модель        
        best_model_tree = model_tree

# Сохраняем наилучшее значение метрики accuracy на валидационных данных
        best_result_tree = result

# Сохраняем глубину дерева решений для наилучшей метрики
        best_depth = depth
        
print("Accuracy лучшей модели на валидационной выборке:", best_result_tree, '\n', 'Глубина дерева решений', best_depth, '\n', 'Accuracy по тестовой выборке:', model_tree.score(features_test, target_test))

metrics(best_model_tree, features_valid, target_valid)

# Вычислим метрику F1
#predicted_valid = best_model_tree.predict(features_valid)
#print('F1-мера:', f1_score(target_valid, predicted_valid))

# Посчитаем площадь под ROC-кривой — AUC-ROC
#probabilities_valid = best_model_tree.predict_proba(features_valid)
#auc_roc = roc_auc_score(target_valid, probabilities_valid[:, 1])
#print('AUC-ROC:', auc_roc)
```

```python
best_model_forest = None
best_result_forest = 0
best_depth_f = 0

# Создаем цикл для выбора количества деревьев
for est in range(78, 82):

# Создаем цикл для выбора глубины дерева решений     
    for depth_f in range (8, 10):
        model_forest = RandomForestClassifier(random_state=12345, n_estimators=est, max_depth=depth_f) 
    
# Обучаем модель на тренировочной выборке
        model_forest.fit(features_train, target_train) 
    
# Расчитываем качество модели на валидационной выборке
        result = model_forest.score(features_valid, target_valid) 
        if result > best_result_forest:
         
# Сохраняем наилучшую модель
            best_model_forest = model_forest 
    
# Сохраняем количество деревьев решений для наилучшей метрики
            estin = est

# Сохраняем глубину дерева решений для наилучшей метрики    
            best_depth_f = depth_f

# Сохраняем наилучшее значение метрики accuracy на валидационных данных    
            best_result_forest = result

print("Accuracy наилучшей модели на валидационной выборке:", best_result_forest, '\n', 'Количество деревьев', estin, '\n', 'Глубина деревьев',  best_depth_f, '\n', 'Accuracy по тестовой выборке:', model_forest.score(features_test, target_test))

metrics(best_model_forest, features_valid, target_valid)

# Вычислим метрику F1
# predicted_valid = best_model_forest.predict(features_valid)
# print('F1-мера:', f1_score(target_valid, predicted_valid))

# Посчитаем площадь под ROC-кривой — AUC-ROC
# probabilities_valid = best_model_forest.predict_proba(features_valid)
# auc_roc = roc_auc_score(target_valid, probabilities_valid[:, 1])
# print('AUC-ROC:', auc_roc)
```

## Борьба с дисбалансом

```python
# Создадим функцию для увеличения выборки
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled

features_upsampled, target_upsampled = upsample(features_train, target_train, 4)

# Посомтрим баланс классов после upsamle
class_frequency = target_upsampled.value_counts(normalize=True) # < напишите код здесь >)
print(class_frequency)
class_frequency.plot(kind='bar')
```
