### Расчет и добавление в таблицу колонки со стоимостью квадратного метра

```python
df['cost_sq_m'] = df['last_price'] / df['total_area']
```

### Добавление в таблицу столбцов с указанием дня недели, месяца и года публикации объявления

```python
days = ['Monday', 'Tuesday', 'Wednsday', 'Thursday', 'Friday', 'Saturday', 'Sunday'] 
weekday = df['first_day_exposition'].dt.dayofweek.apply(lambda x: days[x])
year = df['first_day_exposition'].dt.year.astype(str)
month_name = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
month = df['first_day_exposition'].dt.month.astype(str)
df['weekday'] = weekday
df['month'] = month
df['year'] = year

```

### Добавление в таблицу столбца с указанием этажа расположения квартиры (варианты — первый, последний, другой)

```python
# Создадим функцию для добавления определения типа этажа
def floor_indent (row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'
```

```python
# Применим функцию
df['type_floor'] = df.apply(floor_indent, axis=1)
```

### Добавление в таблицу столбца с соотношением жилой и общей площади

```python
df['liv_ar/tot_ar'] = df['living_area'] / df['total_area']
```

### Добавление в таблицу столбца с отношение площади кухни к общей площади

```python
df['liv_kit/tot_ar'] = df['kitchen_area'] / df['total_area']
```
